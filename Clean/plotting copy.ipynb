```python

import matplotlib.pyplot as plt
import json
import numpy as np
import pandas as pd
```

```python
file_name = "transfer_r15.json"

with open(file_name, "r") as f:
    data = json.load(f)
```

```python
base = data['base']
transfer = data['transfer1']

color = ['navy', 'orange']
```
