```python

import matplotlib.pyplot as plt
import json
import numpy as np
import pandas as pd
```

```python
file_name = "xavier.json"

with open(file_name, "r") as f:
    data = json.load(f)
```

```python

print(data.keys())
```

```python
xavier1 = data["xavier1"]
xavier2 = data["xavier2"]
xavier3 = data["xavier3"]

no_xavier1 = data["no_xavier1"]
no_xavier2 = data["no_xavier2"]
no_xavier3 = data["no_xavier3"]

default = [no_xavier1, no_xavier2, no_xavier3]
xavier = [xavier1, xavier2, xavier3]

color = ['#E41A1C', '#377EB8', '#4DAF4A']
legend = ['Run 1', 'Run 2', 'Run 3']
```

```python
print(len(xavier1), len(xavier2), len(xavier3))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(12,5))
for run, c, l in zip(default, color, legend):
    ax1.semilogy(run, alpha=0.7, color=c, label=l)
ax1.grid(True)
ax1.set_title('Default initialization')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.legend()

for run, c, l in zip(xavier, color, legend):
    ax2.semilogy(run, alpha=0.7, color=c, label=l)
ax2.grid(True)
ax2.set_title('Xavier initialization')
ax2.set_xlabel('Epochs')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
plt.semilogy(xavier1, label="Xavier")
plt.semilogy(xavier2, label="Xavier")
plt.semilogy(xavier3, label="Xavier")

plt.semilogy(no_xavier1, label="Default")
plt.semilogy(no_xavier2, label="Default")
plt.semilogy(no_xavier3, label="Default")

plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Comparing optimization algorithms")
plt.grid()
```
