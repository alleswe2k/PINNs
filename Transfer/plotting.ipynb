```python

import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
file_name = "diff_learning.json"

with open(file_name, "r") as f:
    data = json.load(f)
```

```python

print(data.keys())
```

```python
run1 = data["only_lbfgs"]
run2 = data["only_adam"]
run3 = data["both"]
```

```python
plt.semilogy(run2, label="Only Adam")
plt.semilogy(run3, label="Combining both")
plt.semilogy(run1, label="Only L-BFGS")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Comparing optimization algorithms")
plt.grid()
```

```python
print(len(run1))
print(len(run2))
print(len(run3))
```

```python
x = np.linspace(-3, 3, 100)
y = np.tanh(x)

plt.figure(figsize=(8, 3))

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("The Hyperbolic function Tanh(x)")
plt.grid()
plt.show()
```
