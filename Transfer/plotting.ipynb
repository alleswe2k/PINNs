```python

import matplotlib.pyplot as plt
import json
import os
```

```python
file_name = "transfer2.json"

with open(file_name, "r") as f:
    data = json.load(f)
```

```python

print(data.keys())
```

```python
run1 = data["run1"]
run2 = data["run2"]
```

```python
plt.semilogy(run1, label="Without transfer learning")
plt.semilogy(run2, label="With transfer learning")
plt.legend()
plt.axhline(y=1e-3, color='r', linestyle='--')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Comparing transfer learning")
plt.grid()
```
