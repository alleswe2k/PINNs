```python
import matplotlib.pyplot as plt
import json
from matplotlib import rcParams
```

```python
# Importing the loss data from one of the json files
file_name = "../loss_data/optimization.json"

with open(file_name, "r") as f:
    data = json.load(f)

print(data.keys())
```

```python
adam = data['adam']
lbfgs = data['lbfgs']
both = data['both']

color = ['navy', 'orange']
```

```python
print(f"Total epochs for Adam: {len(adam)}")
print(f"Total epochs for L-BFGS: {len(lbfgs)}")
print(f"Total epochs for both: {len(both)}")
```

```python
# Parameters for changing the font size in the plot
rcParams.update({
    'font.size': 14,
    'axes.titlesize': 14,
    'axes.labelsize': 14,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14,
})
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(12, 5))

ax1.semilogy(adam, color="blue")
ax1.grid(True)
ax1.set_title("Adam")
ax1.set_xlabel("Epochs")
ax1.set_ylabel("Loss")

ax2.semilogy(lbfgs, color="red")
ax2.grid(True)
ax2.set_title("L-BFGS")
ax2.set_xlabel("Epochs")


ax3.semilogy(both, color="green")
ax3.grid(True)
ax3.set_title("Combining both")
ax3.set_xlabel("Epochs")

plt.tight_layout()
plt.show()
```
