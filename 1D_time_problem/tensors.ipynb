```python
import torch
import numpy as np
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
d = torch.linspace(0,1,5).view(-1, 1)
i = torch.ones(4).view(-1, 1) * 2
b = torch.zeros(4).view(-1, 1)

d_len = d.shape[0]
i_len = i.shape[0]
b_len = b.shape[0]

print(d_len, i_len, b_len)

vec = torch.vstack((d, i, b))
print(vec[:d_len])
print(vec[d_len:d_len+i_len])
print(vec[d_len+i_len:d_len+i_len+b_len])
```
