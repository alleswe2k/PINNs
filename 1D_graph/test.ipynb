```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.5
b = 0.5
r = 0.2

theta = np.linspace(0, 2*np.pi, 100)
x = a + r*np.cos(theta)
y = b + r*np.sin(theta)

print(np.max(x))
```

```python
plt.scatter(x, y)
```

```python
nx = -(x - a)/r
ny = -(y - b)/r

plt.scatter(nx, ny)
```
