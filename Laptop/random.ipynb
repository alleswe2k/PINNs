```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 1000)

a1 = lambda x: np.tanh(x)
a2 = lambda x: (x + np.abs(x)) / 2

y1 = a1(x)
y2 = a2(x)
```

```python
plt.figure()
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Tanh activation function")
plt.axvline(x=0, color='black', linestyle='--')

plt.plot(x, y1)
plt.show()

plt.figure()
plt.axvline(x=0, color='black', linestyle='--')
plt.plot(x, y2, color="orange")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("ReLU activation function")
plt.show()
```
