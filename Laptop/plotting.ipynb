```python
import matplotlib.pyplot as plt
import json
import os
```

```python
file_name = "results.json"

with open(file_name, "r") as f:
    data = json.load(f)
```

```python
print(data.keys())
```

```python
no_trans = data["No transfer"]
transfer_freeze = data["transfer_freeze1_10000"]
transfer = data["transfer_no_freeze_10000"]
low_transfer = data["transfer_no_freeze_5000"]
downsample_rate = 100
```

```python
import numpy as np

def moving_average(data, window_size=100):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')
```

```python
no_trans = moving_average(no_trans)
transfer = moving_average(transfer)
```

```python
plt.semilogy(no_trans)
plt.semilogy(transfer)
plt.semilogy(low_transfer)
plt.axhline(y=1e-3, color='r', linestyle='--')
plt.xlabel("Epochs")
plt.ylabel("Loss")
```

```python
plt.semilogy(transfer[::downsample_rate])
```
